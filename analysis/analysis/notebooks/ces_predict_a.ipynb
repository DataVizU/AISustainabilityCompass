```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import statsmodels.api as sm

from ces_model import cal_x
```

```python
df = pd.read_csv('../data/region/human_structure_gdp.csv')
```

```python
df['x'] = cal_x(df['高学历人数'], df['低学历人数'])
```

```python
df.head()
```

```python
df['a'] = df['gdp'] / df['x']
```

```python
df.head()
```

```python
df.to_csv("../data/region/human_structure_gdp_with_a.csv", index=False)
```

```python
df.tail(20)
```

```python
results = {}
models = {}
```

```python
for region in df['地区'].unique():
    df_region = df[df['地区'] == region]

    train_df = df_region[df_region['年份'] != 2021]
    test_df = df_region[df_region['年份'] == 2021]

    model = LinearRegression()

    X_train = train_df[['年份']]
    y_train = train_df['a']
    model.fit(X_train, y_train)

    X_test = test_df[['年份']]
    y_test = test_df['a']
    y_pred = model.predict(X_test)

    error = mean_squared_error(y_test, y_pred)
    results[region] = error

    models[region] = model
```

```python
for region, error in results.items():
    print(f'{region} 的误差为 {error}')
```

```python
results = {}
models = {}
predict = {}
sm_res = {}
```

```python
for region in df['地区'].unique():

    df_region = df[df['地区'] == region]

    train_df = df_region[df_region['年份'] != 2021]
    test_df = df_region[df_region['年份'] == 2021]

    poly = PolynomialFeatures(degree=2)

    X_train = poly.fit_transform(train_df[['年份']])
    X_test = poly.transform(test_df[['年份']])

    model = LinearRegression()

    y_train = train_df['a']
    model.fit(X_train, y_train)

    y_test = test_df['a']
    y_pred = model.predict(X_test)

    error = mean_squared_error(y_test, y_pred)
    results[region] = error
    
    X_train_with_intercept = sm.add_constant(X_train)
    model_with_intercept = sm.OLS(y_train, X_train_with_intercept)
    results_with_intercept = model_with_intercept.fit()
    sm_res[region] = results_with_intercept

    poly = PolynomialFeatures(degree=2)
    train_df = df_region
    X_train = poly.fit_transform(train_df[['年份']])
    model = LinearRegression()
    y_train = train_df['a']
    model.fit(X_train, y_train)

    models[region] = model
    
    year_to_predict = 2023
    X_predict = pd.DataFrame([[year_to_predict]], columns=['年份'])
    X_predict = poly.transform(X_predict)

    predict[region] = model.predict(X_predict)
```

```python
for region, error in results.items():
    print(f'{region} 的误差为 {error}')
```

```python
for region, result in predict.items():
    print(f'{region} 在 2023 年的预测 a 值为 {result[0]}')
```

```python
for region, result in sm_res.items():
    print(region+":")
    print(result.summary())
```

```python
res = []
for region, result in predict.items():
    res.append([region, result[0]])
```

```python
res_df = pd.DataFrame(data=res, columns=["地区", "a"])
res_df.to_csv('../data/region/a_predict_2023.csv', index=False)
```
