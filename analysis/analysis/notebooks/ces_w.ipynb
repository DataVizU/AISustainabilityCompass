```python
import pandas as pd
import sympy as sp
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data_df = pd.read_csv('../data/region/human_structure_gdp_with_a.csv')
```

```python
data_df.head()
```

```python
n = 0.63
k = (n - 1) / n
```

```python
res = []
```

```python
for region in data_df['地区'].unique():
    data = data_df[data_df['地区'] == region]
    data = data[['高学历人数', '低学历人数', 'a', '年份']]
    data = data.values
    for data_i in data:
        h = data_i[0]
        l = data_i[1]
        a = data_i[2]
        x = sp.symbols('x')
        f_h = a * (x ** k + l ** k) ** (1 / k)
        f_prime_h = sp.diff(f_h, x)
        f_l = a * (h ** k + x ** k) ** (1 / k)
        f_prime_l = sp.diff(f_l, x)
        w_h = f_prime_h.subs(x, h)
        w_l = f_prime_l.subs(x, l)
        res.append([region, int(data_i[3]), w_h, w_l])
        
```

```python
print(res)
```

```python
res_pd = pd.DataFrame(data=res, columns=['地区', '年份', '高学历年薪', '低学历年薪'])
```

```python
res_pd.tail(100)
```

```python
res_pd.to_csv("../data/region/年薪.csv", index=False)
```

```python
res_pd["收入差距"] = res_pd["高学历年薪"] / res_pd["低学历年薪"]
```

```python
salary = pd.read_csv('../data/region/年薪.csv', dtype={'年份': int})
```

```python
salary["收入差距"] = salary["高学历年薪"] / salary["低学历年薪"]
```

```python
salary.head()
```

```python
crime = pd.read_csv('../data/region/犯罪率.csv', dtype={'年份': int})
```

```python
crime = pd.melt(crime, id_vars=['地区'], var_name='年份', value_name='犯罪率')
```

```python
crime['年份'] = crime['年份'].astype(int)
```

```python
crime.head()
```

```python
df = pd.merge(salary, crime, on=['地区', '年份'], how='inner')
```

```python
df.head()
```

```python
df = df.dropna(subset=['收入差距', '犯罪率'])
```

```python
results = {}
models = {}
predict = {}
```

```python
for region in df['地区'].unique():
    df_region = df[df['地区'] == region]

    train_df = df_region[df_region['年份'] != 2020]
    test_df = df_region[df_region['年份'] == 2020]

    model = LinearRegression()

    X_train = train_df[['收入差距']]
    y_train = train_df['犯罪率']
    model.fit(X_train, y_train)

    X_test = test_df[['收入差距']]
    y_test = test_df['犯罪率']
    y_pred = model.predict(X_test)

    error = mean_squared_error(y_test, y_pred)
    results[region] = error

    models[region] = model
    
    train_df = df_region
    
    model = LinearRegression()

    X_train = train_df[['收入差距']]
    y_train = train_df['犯罪率']
    model.fit(X_train, y_train)
    
    predict[region] = model.coef_
```

```python
for region, error in results.items():
    print(f'{region} 的误差为 {error}')
```

```python
crime_k = []
```

```python
for region, coef in predict.items():
    print(f'{region} 的回归系数为 {coef}')
    crime_k.append([region, coef[0]])
```

```python
crime_k_df = pd.DataFrame(data=crime_k, columns=['地区', 'k1'])
```

```python
crime_k_df.to_csv("../data/region/犯罪率k1.csv", index=False)
```
